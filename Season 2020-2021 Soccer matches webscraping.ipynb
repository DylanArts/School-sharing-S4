```python
import time
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver

chrome_path = "C:/Program Files/ChromeDriver/chromedriver.exe"
```

```python
link = "https://www.knvb.nl/competities/eredivisie/uitslagen"

data = {
        "HomeTeam": [],
        "AwayTeam": [],
        "Score": []
    }

driver = webdriver.Chrome(chrome_path)
driver.get(link)
for i in driver.find_elements_by_class_name("row"):
    count = 0
    for team in i.find_elements_by_class_name("team"):
        count += 1
        if count == 1:
            data["HomeTeam"].append(team.text)
        else:
            data["AwayTeam"].append(team.text)
        
    data["Score"].append(i.find_element_by_class_name("center").text)
```

```python
eredivisie_matches = pd.DataFrame.from_dict(data)
```

```python
European_Leagues = pd.read_csv("C:/Users/Gebruiker/Desktop/Engineering/Voetbal-Predictor/Code/Goaly/Data/European_Leagues.csv", encoding='iso-8859-1')
European_Leagues.head()
```

```python
European_Leagues['PrimaryLeague'] =  European_Leagues["Competitions"].str.split(" \r\n", n=1, expand=True)[0]
```

```python
European_Leagues
```

```python
baselink = "https://www.skysports.com/"
secondary_part = "-results/2020-21"

data = {
        "HomeTeam": [],
        "HomeScore": [],
        "AwayTeam": [],
        "AwayScore": []
    }

for i in European_Leagues.PrimaryLeague:
    driver.get(baselink + i.replace(" ", "-").lower() + secondary_part)
    time.sleep(1)
    
    for el in driver.find_elements_by_class_name("fixres__item"):
        count = 0
        
        for scores in el.find_elements_by_class_name("matches__teamscores-side"):
            count += 1
            if count == 1:
                data["HomeScore"].append(scores.text)
            else:
                data["AwayScore"].append(scores.text)
            
        count = 0
        for teams in el.find_elements_by_class_name("swap-text__target"):
            count += 1
            if count == 1:
                data["HomeTeam"].append(teams.text)
            else:
                data["AwayTeam"].append(teams.text)
```

```python
matches = pd.DataFrame.from_dict(data)
matches.head()
```
