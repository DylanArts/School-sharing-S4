# Player data webscrape

The following code is code to scrape https://www.whoscored.com/Teams/1. The first script is to scrape only the summary table of a team. The other script is to scrape the detailed table.

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
from selenium.webdriver.common.action_chains import ActionChains
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome_path = "C:\\Users\\artsd\\Drivers\\chromedriver.exe"
```

```python
driver = webdriver.Chrome(chrome_path)

links = ["https://www.whoscored.com/Teams/13/Show/England-Arsenal", 
         "https://www.whoscored.com/Teams/53/Show/Spain-Athletic-Bilbao",
        "https://www.whoscored.com/Teams/764/Show/Netherlands-ADO-Den-Haag",
        "https://www.whoscored.com/Teams/240/Show/Belgium-Anderlecht",
        "https://www.whoscored.com/Teams/5948/Show/Portugal-Arouca",
        "https://www.whoscored.com/Teams/40/Show/Germany-Arminia-Bielefeld",
        "https://www.whoscored.com/Teams/304/Show/France-Paris-Saint-Germain",
        "https://www.whoscored.com/Teams/3473/Show/Turkey-Alanyaspor",
        "https://www.whoscored.com/Teams/80/Show/Italy-AC-Milan",
        ]
baselink = "https://www.whoscored.com"


Club = []
Name = []
data = {
    "CM": [],
    "KG": [],
    "Apps": [],
    "Mins": [],
    "Goals": [],
    "Assists": [],
    "Yel": [],
    "Red": [],
    "SpG": [],
    "PS%": [],
    "AerialsWon": [],
    "MotM": [],
    "Rating": []
}

for i in links:
    driver.get(i)
    leagues = driver.find_element_by_id("teams")

    prefacelinks = []

    for el in leagues.find_elements_by_css_selector("option")[1:]:
        prefacelinks.append(el.get_attribute("value"))

    try:
        for pl in prefacelinks:
            table = driver.find_element_by_id("player-table-statistics-body")
            for tr in table.find_elements_by_css_selector("tr"):
                Club.append(driver.find_element_by_class_name("team-header-name").text)
                Name.append(tr.find_element_by_class_name("player-link").text)

                for td, key in zip(tr.find_elements_by_css_selector("td")[2:], data.keys()):
                    data[key].append(td.text)
            print(baselink + pl)
            driver.get(baselink + pl)
            time.sleep(1)
    except:
        pass
```

```python
data["Club"] = Club
data["Player"] = Name
```

```python
for i in data:
    if len(data[i]) == 2470:
        data[i].append("")
        print("appended")
```

```python
player_data = pd.DataFrame.from_dict(data)
```

```python
df = pd.DataFrame(player_data.Player.str.split('\n').tolist(), columns = ['Number','Name'])
player_data["Name"] = df.Name
player_data["Number"] = df.Number
```

```python
player_data.drop('Player', axis='columns', inplace=True)
```

```python
driver = webdriver.Chrome(chrome_path)


links = ["https://www.whoscored.com/Teams/13/Show/England-Arsenal", 
         "https://www.whoscored.com/Teams/53/Show/Spain-Athletic-Bilbao",
        "https://www.whoscored.com/Teams/764/Show/Netherlands-ADO-Den-Haag",
        "https://www.whoscored.com/Teams/240/Show/Belgium-Anderlecht",
        "https://www.whoscored.com/Teams/5948/Show/Portugal-Arouca",
        "https://www.whoscored.com/Teams/40/Show/Germany-Arminia-Bielefeld",
        "https://www.whoscored.com/Teams/304/Show/France-Paris-Saint-Germain",
        "https://www.whoscored.com/Teams/3473/Show/Turkey-Alanyaspor",
        "https://www.whoscored.com/Teams/80/Show/Italy-AC-Milan",
        ]
baselink = "https://www.whoscored.com"


data_defensive = {
    "CM": [],
    "KG": [],
    "Apps": [],
    "Mins": [],
    "Tackles": [],
    "Inter": [],
    "Fouls": [],
    "Offsides": [],
    "Clear": [],
    "Drb": [],
    "Blocks": [],
    "OwnG": []
}

data_offensive = {
    "CM": [],
    "KG": [],
    "Apps": [],
    "Mins": [],
    "Goals": [],
    "Assists": [],
    "SpG": [],
    "KeyP": [],
    "Drb": [],
    "Fouled": [],
    "Off": [],
    "Disp": [],
    "UnsTch": []
}

data_passing = {
    "CM": [],
    "KG": [],
    "Apps": [],
    "Mins": [],
    "Assists": [],
    "KeyPassPGame": [],
    "AvgPasses": [],
    "Pass_Successes": [],
    "Crosses": [],
    "LongB": [],
    "ThrB": []
}

data_detailed = {
    "CM": [],
    "KG": [],
    "Apps": [],
    "Minutes_Played": [],
    "Shots_Total": [],
    "OutOfBox": [],
    "SixYardBox": [],
    "PenaltyArea": []
}

buttons = ['//*[@id="team-squad-stats-options"]/li[2]/a', 
           '//*[@id="team-squad-stats-options"]/li[3]/a',
           '//*[@id="team-squad-stats-options"]/li[4]/a',
           '//*[@id="team-squad-stats-options"]/li[5]/a']

dictionaries = [data_defensive,
                data_offensive,
                data_passing,
                data_detailed]

ClubaNames = [[[], []],
              [[], []],
              [[], []],
              [[], []]]

for i in links:
    driver.get(i)

    driver.maximize_window()

    leagues = driver.find_element_by_id("teams")

    teams = []

    for el in leagues.find_elements_by_css_selector("option")[1:]:
        teams.append(el.get_attribute("value"))

    for team in teams:
        try:

            for button, d, clubname in zip(buttons, dictionaries, ClubaNames):
                but = WebDriverWait(driver, 3).until(
                    EC.element_to_be_clickable((By.XPATH, button))
                )
                but.click()
                builder = ActionChains(driver)
                builder.move_to_element(but).click(but).perform()

                while True:
                    try:
                        table = driver.find_elements_by_id("player-table-statistics-body")[int(dictionaries.index(d))+1]
                        break
                    except:
                        time.sleep(1)

                for tr in table.find_elements_by_css_selector("tr"):

                    for td, key in zip(tr.find_elements_by_css_selector("td")[2:], d.keys()):
                        d[key].append(td.text)
                    try:
                        clubname[0].append(driver.find_element_by_class_name("team-header-name").text)
                        clubname[1].append(tr.find_element_by_class_name("player-link").text)
                    except:
                        clubname[0].append("")
                        clubname[1].append("")
                        
            driver.get(baselink + team)
            print(baselink+team)
        except:
            pass
```

```python
player_data_defensive = pd.DataFrame.from_dict(data_defensive)
player_data_offensive = pd.DataFrame.from_dict(data_offensive)
player_data_passes = pd.DataFrame.from_dict(data_passing)
player_data_detailed = pd.DataFrame.from_dict(data_detailed)
```
